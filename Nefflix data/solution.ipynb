## 4) in which year highest number of the TV shows & Movies were relesead? show with bar graph.

```python
# create year column
data['Release_Date']=pd.to_datetime(data['Release_Date'])
data['Year']=data['Release_Date'].dt.year
data.head(10)
```

```python
data['Category'].value_counts() # to make sure that are the only two categories
```

```python
year_tot=data.groupby(['Year']).count()['Category']
year_tot
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(year_tot.index,year_tot)
plt.xlabel('Years')
plt.ylabel('Number of movies and TV shows')
plt.show()
```

## 10) maximum duration in a movie

```python
# dataframe with only movies category
mov_set=data.loc[data['Category']=='Movie']
mov_set.head(10)
```

```python
mov_set['time_duration']=mov_set['Duration'].apply(lambda x: x.split(' ')[0])
mov_set.head(5)
```

```python
print('The maximum time duration is '+str(mov_set['time_duration'].max())+' minutes')
```
